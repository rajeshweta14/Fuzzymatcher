```python
import fuzzymatcher
import pandas as pd
from pathlib import *

hospital_accounts = pd.read_csv('hospital_account_info.csv')
hospital_reimbursement = pd.read_csv('hospital_reimbursement.csv')
```

The business scenario is that we want to match up the hospital reibursement information data with hospital customer data. Unfortunately in this scenario we do not have a common ID to join on so we will see if we can use these pythin tools to merge the data together based on a combination of name and address information. 
We will be using fuzzymatcher in this case. For this using the anaconda command prompt I have installed pandas, pathlib and fuzzy_matcher.

Since the columns have differnt names, I am definign account and reimbursement dataframes.

```python
account = ["Facility Name", "Address", "City", "State"]
reimbursement = ["Provider Name", "Provider Street Address", "Provider City",
    "Provider State"]
```

```python
account
```

```python
reimbursement
```

lets now apply fuzzy matching on the account

```python
matched_result = fuzzymatcher.fuzzy_left_join (hospital_accounts,
                                               hospital_reimbursement,
                                               account,
                                               reimbursement,
                                               left_id_col='Account_Num',
                                               right_id_col='Provider_Num')
```

The above statement runs for around 2 mins.


```python
matched_result
```

The matched_results Dataframe contains best_match_score which shows the quality of the link.
Lets arrange the column for in eradable format for the top 5 matches.

```python
cols =[ "best_match_score", "Facility Name", "Provider Name", "Address", "Provider Street Address",
    "Provider City", "City", "Provider State", "State"]
matched_result[cols].sort_values(by=['best_match_score'], ascending=False).head(5)
```

In the above case if we look at the first record the with Best_match_score of 3.09. You can noticed that facility name and provider name, address, city and state are matching. Lets now look at the reverse of the above join.

```python
matched_result[cols].sort_values(by=['best_match_score'], ascending=True).head(5)
```

Now similarly if you look at the value of best_match_score -2.268 you would notice that the facility name, address,city and state are not matching at all this will be a poor match or complete mismatch.

Now lets look at the extreme cases, lets's take a look at some of the matches that might be a little more challenging by looking at scores < 80:

```python
matched_result[cols].query("best_match_score <=.80").sort_values(by=['best_match_score'], ascending=False).head(5)
```

Looking at the above example we can say that fuzzymatcher isa good tool for a medium size data sets. As you start to get to 10000 of rows, it will take a lot of time to compute, so plan accordingly. however the eas of user- especially when working with pandas makes it a great first place to start.
